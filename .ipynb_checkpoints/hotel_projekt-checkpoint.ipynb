# Projekt z zakresu analizy danych

Głównym celem analizy jest stworzenie modelu klasyfikacyjnego, który będzie skoncentrowany na przewidywaniu czy rezerwacja hotelu zostanie odwołana. Analiza ta przyniesie korzyści dla przedsiębiorstwa hotelowego, umożliwiając lepsze zarządzanie dostępnością pokoi oraz optymalizację procesów obsługi rezerwacji.

## opis kolumn

    Booking_ID: unique identifier of each booking
    no_of_adults: Number of adults
    no_of_children: Number of Children
    no_of_weekend_nights: Number of weekend nights (Saturday or Sunday) the guest stayed or booked to stay at the hotel
    no_of_week_nights: Number of week nights (Monday to Friday) the guest stayed or booked to stay at the hotel
    type_of_meal_plan: Type of meal plan booked by the customer:
    required_car_parking_space: Does the customer require a car parking space? (0 - No, 1- Yes)
    room_type_reserved: Type of room reserved by the customer. The values are ciphered (encoded) by INN Hotels.
    lead_time: Number of days between the date of book
    arrival_date: Date of the monthing and the arrival date
    arrival_year: Year of arrival date
    arrival_month: Month of arrival date
    market_segment_type: Market segment designation.
    repeated_guest: Is the customer a repeated guest? (0 - No, 1- Yes)
    no_of_previous_cancellations: Number of previous bookings that were canceled by the customer prior to the current booking
    no_of_previous_bookings_not_canceled: Number of previous bookings not canceled by the customer prior to the current booking
    avg_price_per_room: Average price per day of the reservation; prices of the rooms are dynamic. (in euros)
    no_of_special_requests: Total number of special requests made by the customer (e.g. high floor, view from the room, etc)
    booking_status: Flag indicating if the booking was canceled or not.

```python
corr_values = df.corr()
corr_array = np.array(corr_values)
corr_values = pd.DataFrame(corr_array, columns=corr_values.columns, index=corr_values.index)
corr_values
tril_index = np.tril_indices_from(corr_values)
#trill indices from zwraca indeksy ktore są macierzą dolnie trojkątną
#macierz koerealcji jest symetryczna wiec wystarczy nam gornie trojkatna
# Make the unused values NaNs
for coord in zip(*tril_index):
    corr_values.iloc[coord[0], coord[1]] = np.NaN #ustawiamy na wartosc na `NA` wszystko na dolnie trjkatnej
    
    
    # co to zip(*tril_index)
#####
# tril_index = [(1, 2, 3), (4, 5, 6), (7, 8, 9)]

# # Rozpakowywanie np tuple przy użyciu *
# result = zip(*tril_index)

# # Wynik to zip((1, 2, 3), (4, 5, 6), (7, 8, 9))

#stack() w Pandas służy do przekształcania DataFrame'u z dwoma poziomami indeksu (np. macierzy korelacji) na Series, 
#usuwając jednocześnie jedno z wymiarów indeksu.
#to_frame(): Metoda to_frame() zamienia Series na DataFrame. W tym przypadku,
#po użyciu stack(), mamy teraz jednowymiarową serię korelacji, a to_frame() zamienia ją na jednokolumnowy DataFrame.
#dalej reset index usuwa index co byl i robi normalny 0,1,2,3, itd
#rename zmienia nazwy kolumn
corr_values = (corr_values
               .stack()
               .to_frame()
               .reset_index()
               .rename(columns={'level_0':'feature1',
                                'level_1':'feature2',
                                0:'correlation'}))
corr_values['abs_correlation'] = corr_values.correlation.abs()

#jest komunikat o kolumnach ktore nie sa z wartosciami liczbowymi wiec wylacze komunikaty
```

```python
sns.set_context('talk')
sns.set_style('white')

ax = corr_values.abs_correlation.hist(bins=50, figsize=(12, 8))
ax.set(xlabel='Absolute Correlation', ylabel='Frequency');
```

### Skalowanie

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler #Transform features by scaling each feature to a given range
#szczególnie ważne dla KNN
#Warto pamiętać, że dla niektórych algorytmów skalowanie może poprawić ich wydajność, szczególnie dla tych, które opierają się na odległości między punktami danych, takich jak K-NN czy SVM. Jednakże, drzewa decyzyjne nie są na to wrażliwe.
scaler = RobustScaler()
# scaler = StandardScaler()
# scaler = MinMaxScaler()
# o różnych scalerach: https://medium.com/@onersarpnalcin/standardscaler-vs-minmaxscaler-vs-robustscaler-which-one-to-use-for-your-next-ml-project-ae5b44f571b9
numeric_columns = [
    'no_of_adults',
    'no_of_children',
    'no_of_weekend_nights',
    'no_of_week_nights',
    'lead_time',
    'no_of_previous_cancellations',
    'no_of_previous_bookings_not_canceled',
    'avg_price_per_room',
    'no_of_special_requests'
]
df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

- Accuracy: Dokładność to proporcja poprawnie sklasyfikowanych przypadków (zarówno pozytywnych, jak i negatywnych) wśród wszystkich przypadków w zbiorze danych, mierząca ogólną skuteczność modelu w przewidywaniu poprawnych klas.

- Precision:  To jest miara, która mówi nam, jak wiele z pozytywnie sklasyfikowanych przypadków jest rzeczywiście pozytywnych.
 
- True Positive Rate (Recall/Sensitivity): Czułość to proporcja prawdziwie pozytywnych przypadków wśród wszystkich istniejących przypadków pozytywnych w zbiorze danych, czyli sytuacji, w której klient faktycznie odwołał rezerwację.

- True Negative Rate (Specificity):  Mówi o tym, jak dobrze model radzi sobie z identyfikacją przypadków negatywnych, czyli sytuacji, w której klient nie odwoła rezerwacji (wartość 0).

- F1 Score: F1 Score to miara harmoniczna precyzji i czułości, zapewniająca zrównoważoną ocenę modelu klasyfikacyjnego, szczególnie przy niezbalansowanych klasach.

Confusion Matrix - nie jest to stricte metryka ale jest wizualizacją skuteczności przeprowadzonej klasyfikacji pomocnej w ocenie modelu.

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix, ConfusionMatrixDisplay, precision_recall_fscore_support, precision_score, recall_score, roc_curve, auc, f1_score

def evaluate_metrics(yt, yp, class_probabilities):
    results_pos = {}
    
    # Accuracy
    results_pos['accuracy'] = accuracy_score(yt, yp)
    print(f'Accuracy: {results_pos["accuracy"]:.4f}')
    
    # Precision
    precision = precision_score(yt, yp)
    print(f'Precision: {precision:.4f}')
    
    # Recall (Sensitivity)
    recall = recall_score(yt, yp)
    print(f'Recall (Sensitivity): {recall:.4f}')
    
    # Specificity
    tn, fp, fn, tp = confusion_matrix(yt, yp).ravel()
    specificity = tn / (tn + fp)
    print(f'Specificity: {specificity:.4f}')
    
    # F1 Score
    f1 = 2 * (precision * recall) / (precision + recall)
    print(f'F1 Score: {f1:.4f}')
    
    # ROC Curve
    preds = class_probabilities[:, 1]
    fpr, tpr, threshold = roc_curve(yt, preds)
    roc_auc = auc(fpr, tpr)
    
    plt.title('Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label='AUC = %0.2f' % roc_auc)
    plt.legend(loc='lower right')
    plt.plot([0, 1], [0, 1], 'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()

    return results_pos
```

# Decision Tree

Model drzewa decyzyjnego to narzędzie w dziedzinie uczenia maszynowego, które jest często używane do klasyfikacji i regresji. łówne zalety drzew decyzyjnych to łatwość interpretacji, zdolność do obsługi danych kategorycznych i numerycznych, a także automatyczna selekcja cech. Jednakże, mogą być podatne na przeuczenie, co wymaga zastosowania odpowiednich technik regularyzacyjnych.

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
model_tree = DecisionTreeClassifier(random_state=42)
```

```python
def plot_decision_tree(model, feature_names):
    plt.subplots(figsize=(25, 20)) 
    tree.plot_tree(model, 
                       feature_names=feature_names,  
                       fontsize=11,
                       filled=True,
                       class_names=True)#aby było wiadomo czy chodzi o odwołane
    plt.show()
```

```python
model_tree = DecisionTreeClassifier(random_state=42, max_depth=4)
%time model_tree.fit(X_train, y_train)
preds = model.predict(X_test)
```

W węzłach mamy kilka linijek tekstu. Pierwsza linijka wyświetla warunek decyzyjny, druga to wartość Gini która mierzy impurity (zanieczyszczenie) w danym węźle. Im niższa wartość Gini, tym lepszy podział. Wartość 0 oznacza idealnie czysty węzeł (wszystkie próbki należą do jednej klasy), a wartość 1 oznacza maksymalne zanieczyszczenie (proszki równomiernie rozłożone między różnymi klasami). Trzecia określa liczbę próbek, które trafiają do tego konkretnego węzła. Czwarta linijka Określa liczbę próbek każdej klasy w danym węźle. W przykładzie value=[123, 312] oznacza, że w węźle znajduje się 123 próbek klasy 0 i 312 próbek klasy 1. 

Na podstawie parametru określającego dokładność pracy modelu, takiego jak np. accuracy, możliwe jest przeprowadzenie przeszukiwania siatkowego (grid search) parametrów modelu.

```python
params_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [3, 4, 5, 8, 10, 15, 20],
    'min_samples_leaf': [1, 2, 3, 50]
}
```

- criterion wpływa na to, jak algorytm dokonuje podziałów w drzewie.
- max_depth kontroluje głębokość drzewa.
- min_samples_leaf kontroluje minimalną liczbę próbek w liściu.

```python
grid_search = GridSearchCV(estimator = model_tree, 
                        param_grid = params_grid, 
                        scoring='accuracy',
                        cv = 5, verbose = 1)
grid_search.fit(X_train, y_train.values.ravel())
best_params = grid_search.best_params_
```

```python
#%%time
# You can call `permutation_importance` with the following key arguments:
# - `estimator` the model to be estimated
# - `X` training data X
# - `y` target labels y
# - `n_repeats`, Number of times to permute a feature, each permutation generates an importance value
# Podanie wartości n_jobs może przyspieszyć obliczenia, szczególnie w przypadku dużych zbiorów danych i skomplikowanych modeli.
feature_importances_LL = permutation_importance(estimator=model, X=X_train, y=y_train, n_repeats=5,
                                                      random_state=123, n_jobs=2)
feature_importances_KNN = permutation_importance(estimator=knn, X=X_train, y=y_train, n_repeats=5,
                                                       random_state=123, n_jobs=2)
feature_importances_Tree = permutation_importance(estimator=model_tree, X=X_train, y=y_train, n_repeats=5,
                                                       random_state=123, n_jobs=2)
```

```python
def visualize_feature_importance(importance_array):
    # Sort the array based on mean value
    
    
    #numpy.argsort(a, axis=-1, kind=None, order=None)[source]

# Returns the indices that would sort an array.

# Perform an indirect sort along the given axis using the algorithm specified by the kind keyword.


    sorted_idx = importance_array.importances_mean.argsort()
    # Visualize the feature importances using boxplot
    fig, ax = plt.subplots()
    fig.set_figwidth(16)
    fig.set_figheight(10)
    fig.tight_layout()
    ax.boxplot(importance_array.importances[sorted_idx].T,
               vert=False, labels=X_train.columns[sorted_idx])
    ax.set_title("Permutation Importances (train set)")
    plt.show()
```

`feature_importances_Tree.importances` przedstawiamy jako boxploty posortowane według `feature_importances_Tree.importances_mean`
