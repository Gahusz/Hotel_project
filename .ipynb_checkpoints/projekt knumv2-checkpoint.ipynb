# Wizualizacja

```python
grouped_age_satisfaction = data2.groupby(['Age', 'satisfaction']).size().unstack(level=-1)
#przy grupowaniu warto skorzystać z metody size() która zwraca serię zaweirającą informacje o rozmiarach poszczególnych grup
#size - Return the number of rows if Series. Otherwise return the number of rows times number of columns if DataFrame.
#unstack robi z wartosci w wierszu kolumne
#level okresla w jakim wymiarze obracamy

#tworze nowe kolumny z obliczonymi wartościami procentowymi
grouped_age_satisfaction['Total'] = grouped_age_satisfaction.sum(axis=1) #suma satisfied i nie satisfied dla danego wieku
grouped_age_satisfaction['satisfied (%)'] = (grouped_age_satisfaction['satisfied'] / grouped_age_satisfaction['Total']) * 100
grouped_age_satisfaction['neutral/dissatisfied (%)'] = (grouped_age_satisfaction['neutral or dissatisfied'] / grouped_age_satisfaction['Total']) * 100


plt.figure(figsize=(12, 6))
sns.barplot(x=grouped_age_satisfaction.index, y=grouped_age_satisfaction['satisfied (%)'], color='lightblue', label='Satisfied')
sns.barplot(x=grouped_age_satisfaction.index, y=grouped_age_satisfaction['neutral/dissatisfied (%)'], bottom=grouped_age_satisfaction['satisfied (%)'], color='orange', label='Neutral/Dissatisfied')

plt.xlabel('Wiek')
plt.ylabel('%')
plt.title('Procentowy udział satysfakcji w zależności od wieku')
plt.legend(title='Satisfaction')
plt.xticks(rotation=90) # obrocenie podpisow osi x zeby sie zmiescily
plt.tight_layout() #zmienia polozenie legendy i lepiej wyglada

plt.show()
```

Pozbędę się obserwacji odstających w trzech kolumnach w których wydaje się ich być dużo.

```python
def identify_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = (column < lower_bound) | (column > upper_bound)
    return outliers
#usuwa wszystkie obserwacje odstające w 'rozumieniu boxplota'
```

```python
distance_outliers = identify_outliers(data['Flight.Distance'])
departure_delay_outliers = identify_outliers(data['Departure.Delay.in.Minutes'])
arrive_delay_outliers = identify_outliers(data['Arrival.Delay.in.Minutes'])
```

```python
total_outliers = distance_outliers | departure_delay_outliers | arrive_delay_outliers
print("Ilość usuniętych outliers: " + str(total_outliers.sum()))
```

```python
data = data[~total_outliers]
```

### Regresja logistyczna

```python
from sklearn.model_selection import train_test_split

predictors = [col for col in list(data) if col != 'satisfaction']
#predictors = list(data.columns[:15])
X = data[predictors]
y = data['satisfaction']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=7)
#dzieli 0.8 to zbiór treningowy a testowy to 0.2
```

```python
counts_train = np.bincount(y_train)
print("Liczba 0 w y_train:", counts_train[0])
print("Liczba 1 w y_train:", counts_train[1])

counts_test = np.bincount(y_test)
print("Liczba 0 w y_test:", counts_test[0])
print("Liczba 1 w y_test:", counts_test[1])
```

Widać podobną ilość '0' i '1' zmiennej którą przewiduję w zbiorze treningowym i testowym więc nie ma problemu niezbalansowania klas.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression(max_iter=10000)

%time model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ' + str(accuracy))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

columns_to_standardize = ['Departure.Delay.in.Minutes', 'Arrival.Delay.in.Minutes', 'Flight.Distance']

scaler = StandardScaler()

x_train[columns_to_standardize] = scaler.fit_transform(x_train[columns_to_standardize])
x_test[columns_to_standardize] = scaler.transform(x_test[columns_to_standardize])

model = LogisticRegression(max_iter=10000)

#trenowanie modelu
%time model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print('Accuracy: ' + str(accuracy))
```

standaryzacja dwóch predyktorów z wartościami ciągłymi poprawiła predykcja na 4 miejscu po przecinku

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Not Satisfied', 'Satisfied'], yticklabels=['Not Satisfied', 'Satisfied'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
```

Zwykła regresja logistyczna bez parametrow
- *0.8753729* bez standaryzacji z wyrzuceniem **NA** po prostu
- *0.8743563* zastapienie wartości **NA** medianą, bez wyrzucenia *outliers*
- *0.8718798* ze standaryzacją kolumn numerycznych ciągłych
- *0.8660180* zastapienie wartości **NA** medianą, z wyrzuceniem *outliers*

Moje dodatkowe spostrzeżenia: 
- im mniej daje predyktorów (kolumn) przekazywałem do *predictors* tym mniejsze accuracy
- wyrzucenie outliers w tym przypadku pogarsza jakość modelu
W żadnej wersji nie udało się wejść powyżej 88% accuracy.

### Walidacja krzyżowa

W celu dostrojenia modelu przy jednoczesnym niedopuszczeniu do przeuczenia modelu (nadmiernego dopasowania do zbioru treningowego)
 Sprobuje użyć takiej jak walidacja krzyżowa (K-Fold Cross Validation) która polega na podzieleniu treningowego zbioru danych na kilka podzbiorów w celu symulacji procesu sprawdzania poprawności przewidywań na obserwacjach spoza próbki. Na podstawie parametru określającego dokładność modelu jak np. średni błąd kwadratowy możliwe jest przeprowadzenie przeszukiwania siatkowego (ang. grid search) parametrów modelu.
LogisticRegressionCV - dysponuje klasami estymatorów z wbudowaną walidacją krzyżową. Można jej użyc wraz z parametrem regularyzacji modelu C określającym dokładność wykonywania przeszukiwania siatkowego. (cytat z książki Python w analizie danych Wes McKinney)
